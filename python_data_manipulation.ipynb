```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# information about employees.csv
employees = pd.read_csv('employees.csv')
#print(employees.head())
#print(employees.info())
#print(employees.describe())
#print(employees.isnull().sum())
#print(employees.duplicated().sum())
#print(employees.columns)

# information about company.csv

company = pd.read_csv('company.csv')
#print(company.head())
#print(company.info())
#print(company.describe())
#print(company.isnull().sum())
#print(company.duplicated().sum())
#print(company.columns)


# cleaning format in in employees

employees.columns = (
   employees.columns.str.lower()
   .str.strip()
   .str.replace(' ', '_')
  .str.replace('employeeid', 'employee_id')
  .str.replace('contractleft','contract_left')
)

# cleaning format in company
company.columns =(
     company.columns.str.lower()
     .str.strip()
     .str.replace(' ', '_')
     .str.replace('departmentid', 'department_id')
     .str.replace('employeeid', 'employee_id')

)

# cleaning names and last names in employees
employees['name'] = employees['name'].str.title()
employees['lastname'] = employees['lastname'].str.title()
employees['role'] = employees['role'].str.title()

 # cleaning city, manager, and location to proper format in company
company['city'] = company['city'].str.title()
company['manager'] = company['manager'].str.title()
company['location'] = company['location'].str.title()

# fixing nulls in employees
employees['name'] = employees['name'].fillna('Unknown')
employees['lastname'] = employees['lastname'].fillna('Unknown')
employees['role'] = employees['role'].fillna('Unknown')

# fixing nulls in company
company['city'] = company['city'].fillna('Unknown')
company['manager'] = company['manager'].fillna('Unknown')
company['location'] = company['location'].fillna('Unknown')
company['department'] = company['department'].fillna('Unknown')


# contract left filling NaN with median in employees
employees['contract_left'] = employees['contract_left'].fillna(employees['contract_left'].median())

# cleaning salary with median in company
company['salary'] = company['salary'].fillna(company['salary'].median())

# removing duplicates in employees

employees = employees.drop_duplicates()

duplicate_count = employees.duplicated().sum()

# duplicate in company

company = company.drop_duplicates()

duplicate_count = company.duplicated().sum()

# merging

merged = employees.merge(company, on= 'employee_id', how= 'inner')
results = merged[["employee_id", "name", "department", "role", "salary"]]

# average employee salary per department

average_salary = results.groupby('department')['salary'].mean().round(0)

# employee count per department

employee_count = results.groupby('department')['employee_id'].count()

## saving data back to csv

employees.to_csv("cleaned_employees.csv", index = False)
company.to_csv("cleaned_company.csv", index = False)
results.to_csv("merged_data.csv", index = False)

average_salary.to_csv("average_salary_by_department.csv")
employee_count.to_csv("employee_count_by_department.csv")
```
